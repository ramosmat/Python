```python
#41. Faça um programa que receba o valor de uma dívida e mostre uma tabela com os seguintes dados: valor da dívida, valor
# dos juros, quantidade de parcelas e valor da parcela. Os juros e a quantidade de parcelas seguem a tabela abaixo:
#Quantidade de Parcelas  % de Juros sobre o valor inicial da dívida
#1                         0
#3                         10
#6                         15
#9                         20
#12                        25

div = float(input('Qual o valor da dívida? '))

for i in range(0,13,3): #repetição de 1 parcela até 12 parcelas com intervalo de 3 períodos
    if i == 0:
        i += 1
    
    if 1 <= i <= 2: #Calculos entre 1 e 2 parcelas
        j = 0 #Juros
        v_j = div * j
        div_j = div + (div * j) #Valor da dívida com Juros
        div_p = div_j / i #Valor da parcela
        div_p = round(div_p,2) #Arredondando para 2 decimais
        print('\nDívida: {} | Juros: {} | Valor dos Juros {} | Parcelas: {} | Valor da Parcela: {}'.format(div_j,j,v_j,i,div_p))
    
    if 3 <= i <= 5:
        j = 0.10
        v_j = div * j
        div_j = div + (div * j)
        div_p = div_j / i
        div_p = round(div_p,2)
        print('Dívida: {} | Juros: {} | Valor dos Juros {} | Parcelas: {} | Valor da Parcela: {}'.format(div_j,j,v_j,i,div_p))
        
    if 6 <= i <= 8: #Calculos entre 6 e 8 parcelas
        j = 0.15 #Juros
        v_j = div * j
        div_j = div + (div * j) #Valor da dívida com Juros
        div_p = div_j / i #Valor da parcela
        div_p = round(div_p,2) #Arredondando para 2 decimais
        print('Dívida: {} | Juros: {} | Valor dos Juros {} | Parcelas: {} | Valor da Parcela: {}'.format(div_j,j,v_j,i,div_p))
    
    if 9 <= i <= 11:
        j = 0.20
        v_j = div * j
        div_j = div + (div * j)
        div_p = div_j / i
        div_p = round(div_p,2)
        print('Dívida: {} | Juros: {} | Valor dos Juros {} | Parcelas: {} | Valor da Parcela: {}'.format(div_j,j,v_j,i,div_p))
        
    if i == 12:
        j = 0.25
        v_j = div * j
        div_j = div + (div * j)
        div_p = div_j / i
        div_p = round(div_p,2)
        print('Dívida: {} | Juros: {} | Valor dos Juros {} | Parcelas: {} | Valor da Parcela: {}'.format(div_j,j,v_j,i,div_p))
```

```python
#42. Faça um programa que leia uma quantidade indeterminada de números positivos e conte quantos deles estão nos seguintes
# intervalos: [0-25], [26-50], [51-75] e [76-100]. A entrada de dados deverá terminar quando for lido um número negativo.

c = True
v0_25 = 0
v26_50 = 0
v51_75 = 0
v76_100 = 0
v100 = 0

while c:
    v = int(input('Qual o valor? '))
    
    if v < 0:
        break
    
    elif 0 <= v <= 25:
        v0_25 += 1
    
    elif 26 <= v <= 50:
        v26_50 += 1
        
    elif 51 <= v <= 75:
        v51_75 += 1
        
    elif 76 <= v <= 100:
        v76_100 += 1 
        
    elif v > 100:
        v100 += 1
        
print('\n0 - 25: ',v0_25)
print('26 - 50: ',v26_50)
print('51 - 75: ',v51_75)
print('76 - 100: ',v76_100)
print('> 100: ',v100)
```

```python
#43. Faça um programa que leia o código dos itens pedidos e as quantidades desejadas. Calcule e mostre o valor a ser pago por
# item (preço * quantidade) e o total geral do pedido. Considere que o cliente deve informar quando o pedido deve ser encerrado.

c = True
p100 = 0 #Cachorro Quente
p101 = 0 #Bauru Simples
p102 = 0 #Bauru com Ovo
p103 = 0 #Hambúrguer
p104 = 0 #Cheeseburguer
p105 = 0 #Refrigerante

v100 = 0 #Valor Cachorro Quente
v101 = 0 #Valor Bauru Simples
v102 = 0 #Valor Bauru com Ovo
v103 = 0 #Valor Hambúrguer
v104 = 0 #Valor Cheeseburguer
v105 = 0 #Valor Refrigerante

print('Digite 0 para finalizar.')

while c:
    p = int(input('Código do produto: '))
    
    if p == 0:
        break
        
    elif p == 100:
        p100 += 1
        v100 += 1.20
    
    elif p == 101:
        p101 += 1
        v101 += 1.30
        
    elif p == 102:
        p102 += 1
        v102 += 1.50
        
    elif p == 103:
        p103 += 1
        v103 += 1.20
    
    elif p == 104:
        p104 += 1
        v104 += 1.30
        
    elif p == 105:
        p105 += 1
        v105 += 1
        
    else:
        print('Não é um código válido.\n')

#Script para imprimir apenas os produtos consumidos
if p100 != 0:
    print('{} Cachorro Quente: R$ {}'.format(p100,v100))
if p101 != 0:
    print('{} Bauru Simples: R$ {}'.format(p101,v101))
if p102 != 0:
    print('{} Bauru com Ovo: R$ {}'.format(p102,v102))
if p103 != 0:
    print('{} Hamburguer: R$ {}'.format(p103,v103))
if p104 != 0:
    print('{} Cheeseburguer: R$ {}'.format(p104,v104))
if p105 != 0:
    print('{} Refrigerante: R$ {}'.format(p105,v105))

total = v100 + v101 + v102 + v103 + v104 + v105

print('\nTotal: R$',total)
```

```python
#44. Em uma eleição presidencial existem quatro candidatos. Os votos são informados por meio de código. Os códigos são:
#1 , 2, 3, 4  - Votos para os respectivos candidatos. Você deve montar a tabela ex: 1 - Jose/ 2- João/etc)
#5 - Voto Nulo e 6 - Voto em Branco. Faça um programa que calcule e mostre:
# Para finalizar o conjunto de votos tem-se o valor zero.

c = True

v1 = 0 #Voto Jubscleiton
v2 = 0 #Voto Vick Vapo Rub
v3 = 0 #Voto Abelha
v4 = 0 #Voto Heisenberg
v5 = 0 #Voto Nulo
v6 = 0 #Voto em Branco

print('Digite 0 para finalizar.')
n = 1

while c:
    print('\nVoto numero ',n)
    
    v = int(input('Código do candidato: '))
    
    if v == 0:
        break
        
    elif v == 1:
        v1 += 1
    
    elif v == 2:
        v2 += 1
        
    elif v == 3:
        v3 += 1
        
    elif v == 4:
        v4 += 1
        
    elif v == 5:
        v5 += 1
        
    elif v == 6:
        v6 += 1
    
    n += 1

#O total de votos para cada candidato;
print('\nVotos para Jubscleiton:',v1)
print('Votos para Vick Vapo Rub:',v2)
print('Votos para Abelha:',v3)
print('Votos para Heisenberg:',v4)

#O total de votos nulos;
print('\nVotos nulos:',v5)

#O total de votos em branco;
print('Votos em Branco:',v6)

#A percentagem de votos nulos sobre o total de votos;
total_votos = v1 + v2 + v3 + v4 + v5 + v6
p_vn = (v5/total_votos) * 100

print('Porcentagem de votos nulos: {}%'.format(p_vn))

#A percentagem de votos em branco sobre o total de votos.
p_vb = (v6/total_votos) * 100

print('Porcentagem de votos em branco: {}%'.format(p_vb))

##O vencedor da votação;
can = ['Jubscleiton','Vick Vapo Rub','Abelha','Heisenberg']
vot = [v1,v2,v3,v4]

vencedor_index = vot.index(max(vot))
vencedor = can[vencedor_index]

print('O vencedor da eleição é {} com {} votos'.format(vencedor,max(vot)))
```

```python
#45. Desenvolver um programa para verificar a nota do aluno em uma prova com 10 questões, o programa deve perguntar ao aluno
# a resposta de cada questão e ao final comparar com o gabarito da prova e assim calcular o total de acertos e a nota
# (atribuir 1 ponto por resposta certa). Após cada aluno utilizar o sistema deve ser feita uma pergunta se outro aluno vai
# utilizar o sistema. Após todos os alunos terem respondido informar:

c = True
aluno = 1
aluno_cod = [] #Lista de codigos do aluno
p = 0 #Pontos feitos
p_list = [] #lista de pontos para cada aluno
resp = ['A','B','C','D','E','E','D','C','B','A'] #Gabarito da prova
resp_a = [] #Lista de respostas do aluno

while c:
    aluno_cod.append(aluno) #Coloca o codigo do aluno na lista de codigo
    print('\nAluno',aluno)
    for i in range(1,4): #11
        print('\nQuestão',i)
        r = input('Resposta da questão: ')
        r = r.upper()
        resp_a.append(r) #adiciona a resposta da questao na lista de respostas do aluno
        
    for i in range(0,3): #9
        if resp_a[i] == resp[i]: #compara as respostas do aluno com a do gabarito
            p += 1 #adiciona um ponto para cada resposta certa do aluno
            
    p_list.append(p) #adiciona a quantidade de pontos que cada aluno fez
    q = input('Preencher respostas de outro aluno? ')
    q = q.capitalize()
    if q == 'Sim':
        aluno += 1
        p = 0
        resp_a = [] #zerando a lista de respostas para o proximo aluno preencher
    
    else:
        break

# Maior e Menor Acerto;
maior_index = p_list.index(max(p_list))
maior_a = aluno_cod[maior_index]
maior_posi = p_list[maior_index]

menor_index = p_list.index(min(p_list))
menor_a = aluno_cod[menor_index]
menor_posi = p_list[menor_index]

print('\nA maior quantidade de acertos foi do aluno: {} com {} pontos.'.format(maior_a,maior_posi))
print('A menor quantidade de acertos foi do aluno: {} com {} pontos.'.format(menor_a,menor_posi))

# Total de Alunos que utilizaram o sistema;
print('\nA quantidade de alunos que usaram o sistema foram: ',aluno)

# A Média das Notas da Turma.
med = sum(p_list)/len(p_list)
print('\nA média das notas da turma é: ',med)
```

```python
#46. Em uma competição de salto em distância cada atleta tem direito a cinco saltos. No final da série de saltos de cada atleta,
# o melhor e o pior resultados são eliminados. O seu resultado fica sendo a média dos três valores restantes.
# Você deve fazer um programa que receba o nome e as cinco distâncias alcançadas pelo atleta em seus saltos e depois informe a
# média dos saltos conforme a descrição acima informada (retirar o melhor e o pior salto e depois calcular a média).
# Faça uso de uma lista para armazenar os saltos. Os saltos são informados na ordem da execução, portanto não são ordenados.
# O programa deve ser encerrado quando não for informado o nome do atleta.

c = True
saltos = []

#o while deve perguntar o nome do atleta e atribuir em c,  se n for vazio ou diferente de string, acabar o loop
while c:
    n = input('Insira o nome do atleta: ')
    if len(n) == 0:
        print('Não há mais nenhum atleta.')
        break
    
    else: #Criar lista para receber o valor dos saltos.
        print('\nAtleta: ',n)
        for i in range (1,6):
            s = float(input('Salto {}: '.format(i)))
            saltos.append(s)
                
        #Remover da lista os valores de maior e menor salto
        max_s = max(saltos)
        min_s = min(saltos)
                
        print('\nMelhor salto: {}\nPior Salto: {}'.format(max_s,min_s))
        
        saltos.remove(max_s)
        saltos.remove(min_s)
        
        #calcular a media dos saltos restantes e exibir esta média como resultado final
        med_s = sum(saltos)/len(saltos)
        
        print('Média dos demais saltos:',med_s)
        print('\nResulta final:\n{}: {}'.format(n,med_s))
        
    saltos = []
    print('\nPróximo atleta: ')
```

```python
#47. Em uma competição de ginástica, cada atleta recebe votos de sete jurados. A melhor e a pior nota são eliminadas.
# A sua nota fica sendo a média dos votos restantes. Você deve fazer um programa que receba o nome do ginasta e as notas dos
# sete jurados alcançadas pelo atleta em sua apresentação e depois informe a sua média, conforme a descrição acima informada
# (retirar o melhor e o pior salto e depois calcular a média com as notas restantes). As notas não são informados ordenadas.
# Um exemplo de saída do programa deve ser conforme o exemplo abaixo:

c = True
notas = []


nome = input('Nome do atleta: ')
if len(nome) == 0:
    print('Não há nenhum atleta.')

    
else:
    for i in range (1,8):
        n = float(input('Nota {}:'.format(i)))
        notas.append(n)
        
    max_n = max(notas)
    min_n = min(notas)
        
    notas.remove(max_n)
    notas.remove(min_n)
        
    med_n = sum(notas)/len(notas)

    print('\nResultado final:\nAtleta: {}\nMelhor nota: {}\nPior nota: {}\nMédia: {}'.format(nome,max_n,min_n,med_n))
```

```python
#49. Faça um programa que mostre os n termos da Série a seguir:
# S = 1/1 + 2/3 + 3/5 + 4/7 + 5/9 + ... + n/m.

r = int(input('Quantas repetições tem a série: '))
n = 1
d = 1

for i in range (0,r):
    if i == 0:
        print('Termo {}: {}/{}'.format(i,n,d))
    else:
        n += 1
        d += 2
        print('Termo {}: {}/{}'.format(i,n,d))
```

```python
#50. Sendo H = 1 + 1/2 + 1/3 + 1/4 + ... + 1/N, Faça um programa que calcule o valor de H com N termos.

H = []
n = 1
d = 1

r = int(input('Insira o numero de termos: '))

for i in range (0,r):
    if i == 0:
        v = n/d
        H.append(v)
    
    else:
        d += 1
        v = n/d
        H.append(v)
        
print('O valor da soma é:',sum(H))
```

```python
#51. Faça um programa que mostre os n termos da Série a seguir:
#  S = 1/1 + 2/3 + 3/5 + 4/7 + 5/9 + ... + n/m. 
#Imprima no final a soma da série.

S = []
n = 1
d = 1

r = int(input('Insira o numero de termos: '))

for i in range (0,r):
    if i == 0:
        v = n/d
        v = round(v,2)
        S.append(v)
    
    else:
        n += 1
        d += 2
        v = n/d
        v = round(v,2)
        S.append(v)
        
soma = sum(S)
print('A série é: {}\nA soma da série é: {}'.format(S,soma))
```
