```python
#21. Faça um programa que peça um número inteiro e determine se ele é ou não um número primo.
# Um número primo é aquele que é divisível somente por ele mesmo e por 1.

n = int(input('Insira um número inteiro: '))

if n == 2:
    print('O número é primo.')
    
elif n == 1:
    print('O número não é primo:')

else:
    for i in range(2, n):
        v = n % i #Retorna o resto da divisão
        if v == 0:
            print('O número não é primo.')
            break
        
        else:
            print('O número é primo.')
            break
```

```python
# 22. Altere o programa de cálculo dos números primos, informando, caso o número não seja primo, por quais número ele é divisível.

n = int(input('Insira um número inteiro: '))
d = []

for i in range(1, n + 1):
    v = n % i

    if v == 0: #Se não for primo ele insere o número dividor na lista
        d.append(i)
            
if len(d) > 2:
    print('O número não é primo e é divisivel por:\n{}'.format(d))
    
elif n == 1:
    print('O número não é primo')
    
else:
    print('O número é primo')
```

```python
#23. Faça um programa que mostre todos os primos entre 1 e N sendo N um número inteiro fornecido pelo usuário.
# O programa deverá mostrar também o número de divisões que ele executou para encontrar os números primos.
# Serão avaliados o funcionamento, o estilo e o número de testes (divisões) executados.

numero = int(input("Digite um numero inteiro: "))
if numero == 1 or numero == 2:
    print(
        f"{numero} é primo e foram executadas 0 divisões para descobrir isso"
    )
elif numero % 2 == 0:
    print(
        f"{numero} não é primo e foi executada uma divisão para descobrir isso"
    )
else:
    contador = 1
    primo = True
    for i in range(3, numero, 2):
        contador += 1
        if numero % i == 0:
            primo = False
            break
    if primo:
        print(
            f"{numero} é primo e foram executadas"
            f" {contador} divisões para descobrir isso"
        )
    else:
        print(
            f"{numero} não é primo e foram executadas"
            f" {contador} divisões para descobrir isso"
        )
```

```python
# 26. Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores.
# Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

l = {}
c1 = 0
c2 = 0
c3 = 0

n = int(input('Quantos eleitores votarão: '))

for i in range(1, n + 1):
    v = int(input('Qual candidato você votará: '))
    
    while v < 1 or v > 3:
        v = int(input('O voto é entre 1 e 3: '))
    
    if v == 1:
        c1 += 1
        l['Candidato 1'] = c1
    
    elif v == 2:
        c2 += 1
        l['Candidato 2'] = c2
    
    else:
        c3 += 1
        l['Candidato 3'] = c3

print('A votação terminou com a seguinte contagem:\n',l)
```

```python
#27. Faça um programa que calcule o número médio de alunos por turma. Para isto, peça a quantidade de turmas e a quantidade
# de alunos para cada turma. As turmas não podem ter mais de 40 alunos.

l = {}

t = int(input('Qual a quantidade de turmas: '))

for i in range(1,t + 1):
    q = int(input('Qual a quantidade de alunos para a turma {}: '.format(i)))
    l['Turma' + ' ' + str(i)] = q

med = sum(l.values()) / t

print('A média de alunos é:',med)
```

```python
#31. O Sr. Manoel Joaquim expandiu seus negócios para além dos negócios de 1,99 e agora possui uma loja de conveniências.
# Faça um programa que implemente uma caixa registradora rudimentar. O programa deverá receber um número desconhecido de valores
# referentes aos preços das mercadorias. Um valor zero deve ser informado pelo operador para indicar o final da compra.
# O programa deve então mostrar o total da compra e perguntar o valor em dinheiro que o cliente forneceu, para então calcular e
# mostrar o valor do troco. Após esta operação, o programa deverá voltar ao ponto inicial, para registrar a próxima compra.
# A saída deve ser conforme o exemplo abaixo:

compras = {} #Dicionário vazio
v = 0.01 #Valor inicial para ser diferente de 0
i = 1

while v != 0: #Quando for igual a zero vai parar e mostrar o restultado
    v = input('Valor do produto {}: '.format(i))
    v = float(v)
    compras[i] = v
    print('Produto {}: {} reais'.format(i,v))
    i += 1
        
else:
    print(compras)
    Total = sum(compras.values()) #Soma dos valores no dicionário
    Din = input('Quanto de dinheiro você tem para pagar? ')
    Din = float(Din)
        
if Din < Total:
    print('Com esse valor você não consegue pagar a compra.')
else:
    Troco = Din - Total
    print('Total: {}'.format(Total))
    print('Dinheiro: {}'.format(Din))
    print('O seu troco é de: {}'.format(Troco))
```

```python
#32. Faça um programa que calcule o fatorial de um número inteiro fornecido pelo usuário.
# Ex.: 5!=5.4.3.2.1=120. A saída deve ser conforme o exemplo abaixo:

import numpy as np #biblioteca de funções matematicas

fat = input('Você deseja calcular o fatorial de qual numero? ')
fat = int(fat)

lis = []

for i in range (1, fat+1):
    lis.append(i)
    prod = np.prod(lis)

print('{}! = {} = {}'.format(fat,lis,prod))
#print(prod)
```

```python
#34. Os números primos possuem várias aplicações dentro da Computação, por exemplo na Criptografia.
# Um número primo é aquele que é divisível apenas por um e por ele mesmo. Faça um programa que peça um número inteiro e
# determine se ele é ou não um número primo.

#Pedir o numero
num = input('Qual o número? ')
num = int(num)

x = 0

#Verificar se é impar
    #se for 2, é primo
if num == 2:
    print('O número é primo.')
    
    #se for 1, nao é primo    
elif num == 1:
    print('O número não é primo.')

else: #Rodar um for com range(2,numero informado), se resto da divisao entre o numero e i for zero ele soma 1 à uma variavel x
    #Se x for diferente de zero o numero não é primo
    for i in range(2,num): #Quando colocamos apenas o numero, ele roda o for até o numero anterior
        r = num % i
        if r == 0:
            x += 1
        
    if x != 0:
        print('O número não é primo.')
            
    else:
        print('O número é primo.')
```

```python
#35. Encontrar números primos é uma tarefa difícil. Faça um programa que gera uma lista dos números primos existentes entre
# 1 e um número inteiro informado pelo usuário.

#Fazer um for dentro de for com ranges iguais de 1 até n

n = input('Digite um numero para saber qual ente 1 e ele é primo: ')
n = int(n)

primos = [2]

for i in range (2,n + 1): #varrendo de 1 até o numero inserido
    for j in range (1,n + 1):
        r = i % j #sobra da divisão entre i e j
        if r == 1 and j == 2  and i not in primos:
            primos.append(i)
            
print('Os primos são: ', primos)
```

```python
#37. Uma academia deseja fazer um senso entre seus clientes para descobrir o mais alto, o mais baixo, a mais gordo e o
# mais magro, para isto você deve fazer um programa que pergunte a cada um dos clientes da academia seu código, sua altura
# e seu peso. O final da digitação de dados deve ser dada quando o usuário digitar 0 (zero) no campo código. Ao encerrar o
# programa também deve ser informados os códigos e valores do cliente mais alto, do mais baixo, do mais gordo e do mais magro,
# além da média das alturas e dos pesos dos clientes

n = 1 #numero do cliente
cod_cliente = [] #lista de código dos clientes. Também poderia ser o nome ou qualquer identificação.
alt_cliente = [] #lista de altura dos clientes
pes_cliente = [] #lista de peso dos clientes

cod = True

while cod != 0:
    print('\nN° do cliente: ',n)
    cod = int(input('Insira seu código: '))
    if cod == 0:
        break
    else:
        cod_cliente.append(cod)
        altura = float(input('Qual a altura? '))
        peso = float(input('Qual o peso? '))
        alt_cliente.append(altura)
        pes_cliente.append(peso)
        n += 1

posi_magro = pes_cliente.index(min(pes_cliente)) #index retorna a posição do valor requisitado
posi_gordo = pes_cliente.index(max(pes_cliente))
posi_baixo = alt_cliente.index(min(alt_cliente))
posi_alto = alt_cliente.index(max(alt_cliente))

cod_magro = cod_cliente[posi_magro] #lista[0] -> retorna o item da posição 0 da lista
cod_gordo = cod_cliente[posi_gordo] #cod_cliente[posi_gordo] -> retorna o código do cliente que está na posição posi_gordo
cod_baixo = cod_cliente[posi_baixo]
cod_alto = cod_cliente[posi_alto]

print('\nO cliente mais magro tem código {} e peso {}'.format(cod_magro, min(pes_cliente)))
print('O cliente mais gordo tem código {} e peso {}'.format(cod_gordo, max(pes_cliente)))
print('O cliente mais baixo tem código {} e altura {}'.format(cod_baixo,min(alt_cliente)))
print('O cliente mais alto tem código {} e altura {}'.format(cod_alto,max(alt_cliente)))
```

```python
#40. Foi feita uma estatística em cinco cidades brasileiras para coletar dados sobre acidentes de trânsito.
# Foram obtidos os seguintes dados: Código da cidade; Número de veículos de passeio (em 1999);
# Número de acidentes de trânsito (em 1999). Deseja-se saber:
    #A. Qual o maior e menor índice de acidentes de transito e a que cidade pertence;
    #B. Qual a média de veículos nas cinco cidades juntas;
    #C. Qual a média de acidentes de trânsito nas cidades com menos de 2.000 veículos de passeio.

import random as rd #biblioteca para gerar valores randomicos

cod_cid = [1,2,3,4,5]
num_vei = []
num_aci = []

for i in range(0,5):
    vei = rd.randint(1000,5000) #usando a biblioteca random para randomizar entre 1000 e 5000
    aci = rd.randint(300,4500)
    num_vei.append(vei) #Numero de veiculos
    num_aci.append(aci) #Numero de acidentes

print('Lista veiculos: ',num_vei)
print('Lista acidentes: ',num_aci)

#A.
max_aci = max(num_aci)
min_aci = min(num_aci)
posi_max_aci = num_aci.index(max_aci)
posi_min_aci = num_aci.index(min_aci)

cod_cid_max_aci = cod_cid[posi_max_aci]
cod_cid_min_aci = cod_cid[posi_min_aci]

print('\nA.\nO código da cidade com mais acidentes é: {}, com {} acidentes.'.format(cod_cid_max_aci,max_aci))
print('O código da cidade com menos acidentes é: {} com {} acidentes.'.format(cod_cid_min_aci,min_aci))

#B.
med_vei = sum(num_vei)/len(num_vei)
print('\nB.\nA média de veículos nas cinco cidades é de: ',med_vei)

#C. criar uma lista para acidentes em cidades com menos de 2000 veiculos. Adicionar nessa lista o numero de acidentes caso
# o numero de veiculos seja menor que 2000
aci_2000 = []
cod_2000 = []

for i in range (0,len(num_vei)):
    num_2000 = num_vei[i] #Numero de veiculos na posição i da lista num_vei
    if num_2000 < 2000: #Se o numero de veiculos for menor que 2000
        cod_2000_pos = num_vei.index(num_2000) #Posição do numero de veiculos que for igual a num_2000
        vl_cod_cid = cod_cid[cod_2000_pos] #Valor do codigo da cidade que está na posição do numero de veiculos igual a num_2000
        cod_2000.append(vl_cod_cid) #Criando lista de codigos em cidades que tem o numero de veiculos igual a num_2000
        n_2000_pos = num_vei.index(num_2000) #Posição do numero de veiculos que for igual a num_2000
        vl_aci_2000 = num_aci[n_2000_pos] #Valor de acidentes na cidade que está na posição do numero de veiculos igual num_2000
        aci_2000.append(vl_aci_2000) #Criando lista de acidentes em cidades que tem numero de veiculos igual a num_2000

if len(aci_2000) == 0:
    print('\nC.\nNão há nenhuma cidade com menos que 2000 veículos')
else:
    med_aci_2000 = sum(aci_2000)/len(aci_2000)
    print('\nC.')
    print('As cidades que possuem menos que 2000 veiculos são: {}'.format(cod_2000))
    print('A média de acidentes em cidades que possuem menos de 2000 veiculos é: {}'.format(med_aci_2000))
```
