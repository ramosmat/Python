```python
#5. Faça um programa para a leitura de duas notas parciais de um aluno. O programa deve calcular a média alcançada por aluno e apresentar:
a = float(input('Insira a primeira nota: '))
b = float(input('Insira a segunda nota: '))

notas = [a,b]

med = sum(notas) / len(notas)

#a. A mensagem "Aprovado", se a média alcançada for maior ou igual a sete;
if med >= 7 and med < 10:
    print('Aprovado')
    
#b. A mensagem "Reprovado", se a média for menor do que sete;
elif med < 7:
    print('Reprovado')

#c. A mensagem "Aprovado com Distinção", se a média for igual a dez.
if med == 10:
    print('Aprovado com Distrição')
```

```python
#9. Faça um Programa que leia três números e mostre-os em ordem decrescente.

num = []

x = float(input('Insira um número: '))
y = float(input('Insira outro número: '))
z = float(input('Insira outro número: '))

num.append(x)
num.append(y)
num.append(z)

num.sort() #função para organizar em ordem crescente

num.reverse() #função para inverter as posições de trás para frente

print('Os números em ordem decrescente são: 'num)
```

```python
#10. Faça um Programa que pergunte em que turno você estuda. Peça para digitar M-matutino ou V-Vespertino ou N- Noturno.
# Imprima a mensagem "Bom Dia!", "Boa Tarde!" ou "Boa Noite!" ou "Valor Inválido!", conforme o caso.

print('M = Matutino')
print('V = Vespertino')
print('N = Norturno')

t = str(input('\nInforme seu turno de acordo com as opções acima: '))

while not (t == 'M' or t == 'V' or t == 'N'): #Sintax para verificação de uma entre várias opções
    t = input('Valor inválido. Insira novamente: ')
    
if t == 'M':
    print('Bom dia!')

elif t == 'V':
    print('Boa tarde!')
    
else:
    print('Boa noite!')
```

```python
#11. As Organizações Tabajara resolveram dar um aumento de salário aos seus colaboradores e lhe contrataram para desenvolver o 
# programa que calculará os reajustes.
# Faça um programa que recebe o salário de um colaborador e o reajuste segundo o seguinte critério, baseado no salário atual:
#salários até R$ 280,00 (incluindo) : aumento de 20%
#salários entre R$ 280,00 e R$ 700,00 : aumento de 15%
#salários entre R$ 700,00 e R$ 1500,00 : aumento de 10%
#salários de R$ 1500,00 em diante : aumento de 5%.

nome = input('Qual seu nome? ')

while not nome.isalpha():
    nome = input('Insira o seu nome: ')

sal = input('Qual o seu salário? ')

while not sal.isnumeric():
    sal = input('Insira o seu salário: ')

sal = float(sal)

if sal <= 280:
    au = 0.2
    sal_aj = sal + (sal * au)

elif sal > 280 and sal <=700:
    au = 0.15
    sal_aj = sal + (sal * au)
    
elif sal > 700 and sal <= 1500:
    au = 0.10
    sal_aj = sal + (sal * au)

else:
    au = 0.05
    sal_aj = sal + (sal * au)
    
#Após o aumento ser realizado, informe na tela:
#a. o salário antes do reajuste;
print('O seu salário antes do reajuste era de R$:',sal)

#b. o percentual de aumento aplicado;
print('O ajuste a ser aplicado é de:',au * 100,'%')

#c. o valor do aumento;
print('O valor do aumento foi de: R$', au * sal)

#d. o novo salário, após o aumento.
print('O salário após o reajuste é de: R$',sal_aj)
```

```python
#12. Faça um programa para o cálculo de uma folha de pagamento, sabendo que os descontos são do Imposto de Renda,
# que depende do salário bruto (conforme tabela abaixo) e 3% para o Sindicato e que o FGTS corresponde a 11% do Salário Bruto,
# mas não é descontado (é a empresa que deposita). O Salário Líquido corresponde ao Salário Bruto menos os descontos.
# O programa deverá pedir ao usuário o valor da sua hora e a quantidade de horas trabalhadas no mês.
# Desconto do IR:
# Salário Bruto até 900 (inclusive) - isento
# Salário Bruto até 1500 (inclusive) - desconto de 5%
# Salário Bruto até 2500 (inclusive) - desconto de 10%
# Salário Bruto acima de 2500 - desconto de 20% Imprima na tela as informações.

sal = input('Informe o seu salário por hora: ')

hor = input('Informe as horas trabalhadas no mês: ')

while not sal.isnumeric():
    sal = input('Informe o seu salário apenas com valores numericos: ')

while not hor.isnumeric():
    hor = input('Informe as horas apenas com valores numericos: ')

s = float(float(sal) * float(hor))

if s <= 900:
    ir = 0

elif s > 900 and s <= 1500:
    ir = 0.05

elif s > 1500 and s <= 2500:
    ir = 0.10
    
else:
    ir = 0.20
    
ir = float(ir)

d_ir = float(s * ir)

d_sind = float(s * 0.03)

d_inss = float(s * 0.1)

d_tot = d_ir + d_sind + d_inss

s_l = s - d_tot

print('IR: R${}\nINSS: R${}\nSindicato: R${}\nDesconto total: R${}\nSalário líquido: R${}'.format(d_ir,d_inss,d_sind,d_tot,s_l))
```

```python
#14. Faça um programa que lê as duas notas parciais obtidas por um aluno numa disciplina ao longo de um semestre,
# e calcule a sua média. A atribuição de conceitos obedece à tabela abaixo:

#Média de Aproveitamento  Conceito
#  Entre 9.0 e 10.0        A
#  Entre 7.5 e 9.0         B
#  Entre 6.0 e 7.5         C
#  Entre 4.0 e 6.0         D
#  Entre zero e 4.0        E

# O algoritmo deve mostrar na tela as notas, a média, o conceito correspondente e a mensagem “APROVADO” se o conceito
# for A, B ou C ou “REPROVADO” se o conceito for D ou E.

n = []

a = input('Insira a primeira nota: ')
b = input('Insira a segunda nota: ')

a = float(a)
b = float(b)

n.append(a)
n.append(b)

med = (sum(n) / len(n))

if med >= 9 and med <= 10:
    con = 'A'
    
if med >= 7.5 and med < 9:
    con = 'B'
    
if med >= 6 and med < 7.5:
    con = 'C'
    
if med >= 4 and med < 6:
    con = 'D'
    
if med >= 0 and med < 4:
    con = 'E'
    
if con == 'A' or con == 'B' or con == 'C':
    print('Notas: {}\nMédia: {}\nConceito: {}\nAprovado!'.format(n,med,con))

else:
    print('Notas: {}\nMédia: {}\nConceito: {}\nReprovado!'.format(n,med,con))
```

```python
#15. Faça um Programa que peça os 3 lados de um triângulo. O programa deverá informar se os valores podem ser um triângulo.
# Indique, caso os lados formem um triângulo, se o mesmo é: equilátero, isósceles ou escaleno.
# Dicas:
# Três lados formam um triângulo quando a soma de quaisquer dois lados for maior que o terceiro; a+b>c a+c>b b+c>a
# Triângulo Equilátero: três lados iguais;
# Triângulo Isósceles: quaisquer dois lados iguais;
# Triângulo Escaleno: três lados diferentes;

import math

a = input('Valor do primeiro lado: ')
b = input('Valor do segundo lado: ')
c = input('Valor do terceiro lado: ')

while not a.isnumeric():
    a = input('Insira um valor numerico para o primeiro lado: ')

while not b.isnumeric():
    b = input('Insira um valor numerico para o segundo lado: ')

while not c.isnumeric():
    c = input('Insira um valor numerico para o terceiro lado: ')

a = float(a)
b = float(b)
c = float(c)

if math.copysign(a - b, 0) < c < a + b and math.copysign(a - c, 0) < b < a + c and math.copysign(b - c, 0) < a < b + c:
    print('É um triângulo.')
    
#math.copysing(x,y) - Esta função serve para copiar o sinal de y para x: math.copysing(-1, +0) = +1

    if a == b == c:
        print('É um triângulo equilátero.')
    
    elif a == b or a == c or b == c:
        print('É um triângulo isósceles.')
        
    elif a != b != c:
        print('É um triângulo escaleno.')
    
else:
    print('Não é um triângulo.')
```
