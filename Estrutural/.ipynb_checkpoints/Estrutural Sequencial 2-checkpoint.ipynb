```python
#14. João Papo-de-Pescador, homem de bem, comprou um microcomputador para controlar o rendimento diário de seu trabalho.
# Toda vez que ele traz um peso de peixes maior que o estabelecido pelo regulamento de pesca do estado de São Paulo (50 quilos)
# deve pagar uma multa de R$ 4,00 por quilo excedente. João precisa que você faça um programa que leia a variável peso
# (peso de peixes) e calcule o excesso. Gravar na variável excesso a quantidade de quilos além do limite e na variável multa o
# valor da multa que João deverá pagar. Imprima os dados do programa com as mensagens adequadas.

pl = 50

pr = float(input('Qual é o peso dos peixes? '))

dif = pr - pl

if dif <= 0:
    print('Você não pagará multa.')

else:
    m = dif * 4
    m = float(m)
    print('Você pagará uma multa de R$',m)
```

```python
#15. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.
# Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para o Imposto de Renda,
# 8% para o INSS e 5% para o sindicato, faça um programa que nos dê:

s = float(input('Qual o seu salário por hora? '))

h = float(input('Quantas horas você trabalha no mês? '))

#a. salário bruto.

sb = s * h

print('O valor do salário bruto é: R$',sb)

#b. quanto pagou ao INSS.
inss = sb * 0.08

print('O valor pago ao INSS é: R$',inss)

#c. quanto pagou ao sindicato.
sind = sb * 0.05

print('O valor pago ao sindicato é: R$',sind)

#d. quanto pagou ao imposto de renda.
ir = sb * 0.11

print('O valor pago ao IR é: R$',ir)

#e. calcule os descontos e o salário líquido, conforme a tabela abaixo:
d = ir + sind + inss

sl = sb - d

print('O valor dos descontos é: R$',d,'\nO valor do salário liquido é: R$',sl)
```

```python
#16. Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada.
# Considere que a cobertura da tinta é de 3 metros quadrados para cada 1 litro e que a tinta é vendida em latas de 18 litros,
# que custam R$ 80,00. Informe ao usuário a quantidades de latas de tinta a serem compradas e o preço total.

a = int(input('Qual a área que será pintada: '))

li = float(a / 3)  #Esta operação é para descobrir quantos litros serão usados (m²/(m²/L) = L)

nl = float(li / 18) #Esta operação é para descobrir quantas latas serão usadas (L/(L/lata) = lata)

#Arredondando o numero de latas para cima:
import math
nl = math.ceil(nl)

P = nl * 80 #Preço a se pagar pelas latas

#-----------------
P = round(P,2)

print('O numero de latas a se comprar é:',nl,'\nO valor a se pagar é: R$',P)
```

```python
#18. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps),
# calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

ta = float(input('Qual o tamanho do arquivo que será baixado em MB? ')) #tamanho em megabytes
vel = float(input('Qual a velocidade da internet em Mbps? ')) #velocidade em megabits por segundo

seg = ta / vel #tempo de download em segundos

mi = int(seg / 60) #tempo de download em minutos

seg2 = seg%60 #restante do tempo de download em segundos

import math
seg2 = math.ceil(seg2)

print('Para baixar um arquivo de',ta,'MB com uma internet de',vel,'Mbps, levará aproximadamente',mi,'minutos e',seg2,'segundos')
```
