```python
#1. Faça um programa para imprimir:
#    1
#    2   2
#    3   3   3
#    .....
#    n   n   n   n   n   n  ... n
# para um n informado pelo usuário. Use uma função que receba um valor n inteiro e imprima até a n-ésima linha.

p = int(input('Insira um valor: '))

def rep(n):
    for i in range (1, n+1):
        for j in range(1, i+1):
            print(i, end = ' ')
        
        print('\n')
    
rep(p)
```

```python
#5. Faça um programa com uma função chamada somaImposto. A função possui dois parâmetros formais: taxaImposto, que é a
# quantia de imposto sobre vendas expressa em porcentagem e custo, que é o custo de um item antes do imposto.
# A função “altera” o valor de custo para incluir o imposto sobre vendas.

def somaImposto(taxaImposto,custo):
    tp = taxaImposto/100
    custop = custo + (custo * tp)
    print('O custo do produto com impostos é:',custop)
    
taxa = float(input('Qual o valor do imposto? '))
c = float(input('Qual o valor do produto sem impostos?'))

somaImposto(taxa,c)
```

```python
#7. Faça um programa que use a função valorPagamento para determinar o valor a ser pago por uma prestação de uma conta.
# O programa deverá solicitar ao usuário o valor da prestação e o número de dias em atraso e passar estes valores para a função
# valorPagamento, que calculará o valor a ser pago e devolverá este valor ao programa que a chamou. O programa deverá então
# exibir o valor a ser pago na tela. Após a execução o programa deverá voltar a pedir outro valor de prestação e assim continuar
# até que seja informado um valor igual a zero para a prestação. Neste momento o programa deverá ser encerrado, exibindo o
# relatório do dia, que conterá a quantidade e o valor total de prestações pagas no dia.

# O cálculo do valor a ser pago é feito da seguinte forma. Para pagamentos sem atraso, cobrar o valor da prestação.
# Quando houver atraso, cobrar 3% de multa, mais 0,1% de juros por dia de atraso.

vp = 0 #valor de prestações
qp = 0 #quantidade de prestações

def valorPagamento(pres,atr): #recebe prestação e atraso
    global vp,qp
    
    if atr == 0:
        pag = pres
        
    else:
        pag = pres + (pres*0.03) + (pres*0.001*atr)
        
    print('O valor da prestação é:',pag)
    vp += pag
    qp += 1

print('Digite zero para encerrar.')
print('')

while True:
    v = float(input('Valor da prestação: '))
    
    while v < 0:
        print('Valor da prestação deve ser 0 ou maior.')
        print('')
    
    if v == 0:
        print('\nPrograma encerrado.')
        print('\nValor total das prestações do dia:',vp)
        print('Quantidade de prestações do dia:',qp)
        
        vp = 0
        qp = 0
        break
        
    else:
        d = int(input('Dias atrasados: '))
        
        while d < 0:
            print('Valor do atraso deve ser 0 ou maior.')
            print('')
        
    valorPagamento(v,d)
```

```python
#9. Reverso do número. Faça uma função que retorne o reverso de um número inteiro informado. Por exemplo: 127 -> 721.

def reverso(n):
    x = str(n)
    a = x[:: -1]
    return a

n = int(input('Digite um valor informado: '))
print('O reverso do numero',n,'é',reverso(n))
```

```python
#10. Jogo de Craps. Faça um programa de implemente um jogo de Craps.
# O jogador lança um par de dados, obtendo um valor entre 2 e 12. Se, na primeira jogada, você tirar 7 ou 11, você é um
# "natural" e ganhou. Se você tirar 2, 3 ou 12 na primeira jogada, isto é chamado de "craps" e você perdeu.
# Se, na primeira jogada, você fez um 4, 5, 6, 8, 9 ou 10,este é seu "Ponto". Seu objetivo agora é continuar jogando os dados
# até tirar este número novamente. Você perde, no entanto, se tirar um 7 antes de tirar este Ponto novamente.

import random as rd
c = 1

def rolar():
    d1 = rd.randint(1,6)
    d2 = rd.randint(1,6)
    r = d1 + d2
    return r

p = rolar() #pontos

if p == 7 or p == 11:
    print('Parabens, você rolou um',p,'e ganhou o jogo!')
    
elif p == 2 or p == 3 or p == 12:
    print('Que pena, você perdeu rolou um',p,'e perdeu o jogo.')
    
else:
    while True:
        s = rolar()
        c += 1
        if s == 7:
            print('Que pena, você rolou os dados',c,'vezes, tirou 7 e perdeu.')
            break
        
        elif s == p:
            print('Parabens, você rolou os dados',c,'vezes e ganhou o jogo antes de tirar 7!')
            break
```
