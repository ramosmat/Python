```python
#16. Faça um programa que calcule as raízes de uma equação do segundo grau, na forma ax2 + bx + c.
# O programa deverá pedir os valores de a, b e c e fazer as consistências, informando ao usuário nas seguintes situações:
# Se o usuário informar o valor de A igual a zero, a equação não é do segundo grau e o programa não deve fazer pedir os demais
#valores, sendo encerrado;

a = input('Insira o valor de a: ')

if a == '0':
    print('Não é uma equação de segundo grau.')
    
else:
    a = float(a)
    b = input('Insira o valor de b: ')
    c = input('Insira o valor de c: ')
    
    b = float(b)
    c = float(c)
    
    d = (b ** 2) - 4 * a * c

# Se o delta calculado for negativo, a equação não possui raizes reais. Informe ao usuário e encerre o programa;

    if d < 0:
        print('A equação não possui raizes reais.')        
        
# Se o delta calculado for igual a zero a equação possui apenas uma raiz real; informe-a ao usuário;
    elif d == 0:
        x1 = (-b + (d ** 0.5))/(2 * a)
        print('A equação tem apenas uma raiz, de valor: {}'.format(x1))
        
# Se o delta for positivo, a equação possui duas raiz reais; informe-as ao usuário;
    elif d > 0:
        x1 = (-b + (d ** 0.5))/(2 * a)
        x2 = (-b - (d ** 0.5))/(2 * a)
        print('A equação tem duas raizes, de valores: {} e {}'.format(x1,x2))
```

```python
#20. Faça um Programa para leitura de três notas parciais de um aluno. O programa deve calcular a média alcançada por aluno e
# presentar:
# A mensagem "Aprovado", se a média for maior ou igual a 7, com a respectiva média alcançada;
# A mensagem "Reprovado", se a média for menor do que 7, com a respectiva média alcançada;
# A mensagem "Aprovado com Distinção", se a média for igual a 10.

n = []

a = input('Insira a primeira nota: ')
b = input('Insira a segunda nota: ')
c = input('Insira a terceira nota: ')

a = float(a)
b = float(b)
c = float(c)

n.append(a)
n.append(b)
n.append(c)

med = (sum(n) / len(n))

med = round(med,2)

if 7 <= med < 10:
    print('Aprovado com média: {}'.format(med))
    
elif med <= 7:
    print('Reprovado com média: {}'.format(med))
    
elif med == 10:
    print('Aprovado com Distinção! Média: {}'.format(med))
```

```python
#21. Faça um Programa para um caixa eletrônico. O programa deverá perguntar ao usuário a valor do saque e depois informar
# quantas notas de cada valor serão fornecidas. As notas disponíveis serão as de 1, 5, 10, 50 e 100 reais.
#O valor mínimo é de 10 reais e o máximo de 600 reais.
# O programa não deve se preocupar com a quantidade de notas existentes na máquina.
# Exemplo 1: Para sacar a quantia de 256 reais, o programa fornece duas notas de 100, uma nota de 50,
# uma nota de 5 e uma nota de 1;
# Exemplo 2: Para sacar a quantia de 399 reais, o programa fornece três notas de 100, uma nota de 50,
# quatro notas de 10, uma nota de 5 e quatro notas de 1.

valor = int(input('Digite o valor para saque: '))

while valor < 10 or valor > 600:
    valor = int(input('Valor inválido! Digite um valor maior que R$10 e menor que R$600.'))

n100 = int((valor - (valor % 100)) / 100)
n50 = int((valor % 100) / 50)
n10 = int(((valor % 50) / 10))
n5 = int((valor % 10) / 5)
n1 = valor % 5

if n100 == 1:
    print(f'{n100} nota de R$100')

elif n100 > 1:
    print(f'{n100} notas de R$100')

if n50 == 1:
    print(f'{n50} nota de R$50')

elif n50 > 1:
    print(f'{n50} notas de R$50')

if n10 == 1:
    print(f'{n10} nota de R$10')

elif n10 > 1:
    print(f'{n10} notas de R$10')

if n5 == 1:
    print(f'{n5} nota de R$5')

elif n5 > 1:
    print(f'{n5} notas de R$5')
    
if n1 == 1:
    print(f'{n1} nota de R$1')
    
elif (n1 > 1):
    print(f'{n1} notas de R$1')
```

```python
#24. Faça um Programa que leia 2 números e em seguida pergunte ao usuário qual operação ele deseja realizar.
# O resultado da operação deve ser acompanhado de uma frase que diga se o número é:
# par ou ímpar;
# positivo ou negativo;
# inteiro ou decimal.

n1 = input('Insira um número: ')

n2 = input('Insira outro número: ')

n1 = float(n1)

n2 = float(n2)

op = input('Qual operação você deseja fazer? ')

if op == 'adição' or op =='mais':
    n = n1 + n2
    op = 'adição'
    
elif op == 'subtração' or op == 'menos':
    n = n1 - n2
    op = 'subtração'
    
elif op == 'multiplicação' or op == 'vezes':
    n = n1 * n2
    op = 'multiplicação'

elif op == 'divisão' or op == 'dividir':
    n = n1 / n2
    op = 'divisão'
    
if n % 2 == 0:
    a = 'par'

else:
    a = 'impar'
    
if n >= 0:
    b = 'positivo'
    
else:
    b = 'negativo'
    
if n % 1 == 0:
    c = 'inteiro'
    
else:
    c = 'decimal'
    
print('Os números inseridos são {} e {}.\nO resultado da operação de {} é igual à {}.\nÉ um número {}, {} e {}'
     .format(n1,n2,op,n,a,b,c))
```

```python
#25. Faça um programa que faça 5 perguntas para uma pessoa sobre um crime. As perguntas são:
# "Telefonou para a vítima?"
# "Esteve no local do crime?"
# "Mora perto da vítima?"
# "Devia para a vítima?"
# "Já trabalhou com a vítima?"
# O programa deve no final emitir uma classificação sobre a participação da pessoa no crime.
# Se a pessoa responder positivamente a 2 questões ela deve ser classificada como "Suspeita", entre 3 e 4 como "Cúmplice"
# e 5 como "Assassino". Caso contrário, ele será classificado como "Inocente".

ev = []

a = input('Telefonou para a vítima? ')
b = input('Esteve no local do crime? ')
c = input('Mora perto da vítima? ')
d = input('Devia para a vítima? ')
e = input('Já trabalhou com a vítima? ')

ev.append(a)
ev.append(b)
ev.append(c)
ev.append(d)
ev.append(e)

j = 0

for i in ev:
    if i == 'sim' or i == 'Sim':
        j = j + 1
        
if j == 2:
    print('Suspeito(a)')
    
elif 3 <= j <= 4:
    print('Cúmplice')
    
elif j == 5:
    print('Assassino')
    
else:
    print('Inocente')
```

```python
#26. Um posto está vendendo combustíveis com a seguinte tabela de descontos:
#Álcool:
#  até 20 litros, desconto de 3% por litro
#  acima de 20 litros, desconto de 5% por litro
#Gasolina:
#  até 20 litros, desconto de 4% por litro
#  acima de 20 litros, desconto de 6% por litro

#Escreva um algoritmo que leia o número de litros vendidos, o tipo de combustível (codificado da seguinte forma: A-álcool,
# G-gasolina), calcule e imprima o valor a ser pago pelo cliente sabendo-se que o preço do litro da gasolina é R$ 2,50
# o preço do litro do álcool é R$ 1,90.

t = input('Insira A para Álcool e G para Gasolina: ')

l = input('Insira a quantidade de combustível em litros: ')

l = float(l)

pg = 2.5 #R$/L

pa = 1.9 #R$/L

if t == 'A':
    if l <= 20:
        pf = (pa * l) - ((pa * l) * 0.03)
    
    else:
        pf = (pa * l) - ((pa * l) * 0.05)
    
    t = 'Álcool'

if t == 'G':
    if l <= 20:
        pf = (pg * l) - ((pg * l) * 0.04)
    
    else:
        pf = (pg * l) - ((pg * l) * 0.06)
        
    t = 'Gasolina'
        
print('Por {} de {} você pagará R${}'.format(l,t,pf))
```

```python
#27. Uma fruteira está vendendo frutas com a seguinte tabela de preços:
#                       Até 5 Kg           Acima de 5 Kg
# Morango         R$ 2,50 por Kg          R$ 2,20 por Kg
# Maçã            R$ 1,80 por Kg          R$ 1,50 por Kg
# Se o cliente comprar mais de 8 Kg em frutas ou o valor total da compra ultrapassar R$ 25,00,
# receberá ainda um desconto de 10% sobre este total. Escreva um algoritmo para ler a quantidade (em Kg) de morangos e a
# quantidade (em Kg) de maças adquiridas e escreva o valor a ser pago pelo cliente.

qmo = input('Qual quantidade de morangos você irá comprar em Kg? ')
qma = input('Qual quantidade de maçãs você irá comprar em Kg? ')

qmo = float(qmo)
qma = float(qma)

if qmo <= 5:
    pmo = qmo * 2.5

else:
    pmo = qmo * 2.2
    
if qma <= 5:
    pma = qma * 1.8

else:
    pma = qma * 1.5
    
if qmo + qma > 8 or pmo + pma > 25:
    print('Você ainda tem um desconto de 10%')
    pf = (pmo + pma) -((pmo + pma) * 0.1)
    
else:
    pf = pmo + pma

print('A compra custará: R${}'.format(pf))
```

```python
#28. O Hipermercado Tabajara está com uma promoção de carnes que é imperdível. Confira:
#                       Até 5 Kg           Acima de 5 Kg
# File Duplo      R$ 4,90 por Kg          R$ 5,80 por Kg
# Alcatra         R$ 5,90 por Kg          R$ 6,80 por Kg
# Picanha         R$ 6,90 por Kg          R$ 7,80 por Kg

# Para atender a todos os clientes, cada cliente poderá levar apenas um dos tipos de carne da promoção,
# porém não há limites para a quantidade de carne por cliente. Se compra for feita no cartão Tabajara o
# cliente receberá ainda um desconto de 5% sobre o total da compra. Escreva um programa que peça o tipo e
# a quantidade de carne comprada pelo usuário e gere um cupom fiscal, contendo as informações da compra:
# tipo e quantidade de carne, preço total, tipo de pagamento, valor do desconto e valor a pagar.

t = input('Qual a carne desejada? Temos File Duplo, Alcatra e Picanha: ')

t = t.capitalize()

q = input('Qual a quantidade de carne desejada? ')

q = float(q)

fp = input('Você pagará com cartão Tabajara? ')

fp = fp.capitalize()

if t == 'File duplo':
    if q <= 5:
        p = q * 4.9
    else:
        p = q * 5.8

if t == 'Alcatra':
    if q <= 5:
        p = q * 5.9
    else:
        p = q * 6.8
        
if t == 'Picanha':
    if q <= 5:
        p = q * 6.9
    else:
        p = q * 7.8
        
if fp == 'Sim':
    p = p - (p * 0.05)
    tp = 'Cartão Tabajara'
    
else:
    tp = 'Dinheiro, credito ou debito'
    
print('Carne: {}\nSubtotal: R${}\nTipo de pagamento: {}\nTerá desconto: {}\nTotal: {}'.format(t,p,tp,fp,p))
```
