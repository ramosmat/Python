```python
#4. Faça um Programa que peça as 4 notas bimestrais e mostre a média.

print('para calcular a média das 4 notas no bimestre, insira cada uma a seguir.')

n1 = input('Nota 1º bimestre: ')
while not n1.isnumeric():
    print('Tem que ser um numero! Insira novamente a primeira nota:')
    n1 = input()

n2 = input('Nota 2º bimestre: ')
while not n2.isnumeric():
    print('Tem que ser um numero! Insira novamente a segunda nota:')
    n2 = input()

n3 = input('Nota 3º bimestre: ')
while not n3.isnumeric():
    print('Tem que ser um numero! Insira novamente a terceira nota:')
    n3 = input()

n4 = input('Nota 4º bimestre: ')
while not n4.isnumeric():
    print('Tem que ser um numero! Insira novamente a quarta nota:')
    n4 = input()

n1 = int(n1)
n2 = int(n2)
n3 = int(n3)
n4 = int(n4)

soma_ns = n1 + n2 + n3 + n4

media = soma_ns/4

print('A média das notas é: {}'.format(media))
```

```python
#4.1 Faça um Programa que peça as 4 notas bimestrais e mostre a média.

print('para calcular a média das 4 notas no bimestre, insira cada uma a seguir.')

n1 = input('Nota 1º bimestre: ')
while not n1.isnumeric():
    print('Tem que ser um numero! Insira novamente a primeira nota:')
    n1 = input()

n2 = input('Nota 2º bimestre: ')
while not n2.isnumeric():
    print('Tem que ser um numero! Insira novamente a segunda nota:')
    n2 = input()

n3 = input('Nota 3º bimestre: ')
while not n3.isnumeric():
    print('Tem que ser um numero! Insira novamente a terceira nota:')
    n3 = input()

n4 = input('Nota 4º bimestre: ')
while not n4.isnumeric():
    print('Tem que ser um numero! Insira novamente a quarta nota:')
    n4 = input()

n1 = int(n1)
n2 = int(n2)
n3 = int(n3)
n4 = int(n4)

ns = [n1,n2,n3,n4]

qnt_ns = len(ns)

soma_ns = n1 + n2 + n3 + n4

media = soma_ns / qnt_ns

print('A média das notas é: {}'.format(media))
```

```python
#5. Faça um Programa que converta metros para centímetros.

print('Vamos transformar metros para centimetros!')

m = input('Insira um número com valor em metros: ')

while not m.isnumeric():
    print('Insira um núemero seu imbecil!')
    m = input()

m = int(m)
cm = m*100

print('O valor {} em metros é equivalente à {} em centimetros.'.format(m,cm))
```

```python
#6 Faça um Programa que peça o raio de um círculo, calcule e mostre sua área.

import math

print('Vamos calcular a área de um círculo.')

r = input('Insira o valor do raio: ')

while not r.isnumeric():
    print('Insira um numero caceta!')
    r = input()

r = int(r)

uni = input('Qual a unidade do valor do raio? ')

while not uni.isalpha() or uni != 'metros' and uni != 'centimetros':
        print('A unidade deve ser: metros ou centimetros.')
        uni = input()

A = math.pi*r**2

print('A área do circulo é {} {}.'.format(A, uni))
```

```python
#7 Faça um Programa que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.

print('Insira o valor do lado de um quadrado para saber o valor do dobro da área!')

l = input()

while not l.isnumeric():
    print('Insira um numero: ')
    l = input()

l = int(l)

uni = input('Qual a unidade do valor inserido? ')

while not uni.isalpha or uni != 'm' and uni != 'cm':
    print('A unidade deve ser: m ou cm.')
    uni = input()

uni = str(uni)

dA = 2*(l*l)

print('O valor do dobro da área do quadrdado é {} {}²'.format(dA, uni))
```

```python
# 8 Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.
## Calcule e mostre o total do seu salário no referido mês.

print('Vamos estimar o quanto você pode ganhar no mês?')

nome = input('Qual é o seu nome? ') #Nome da pessoa

while not nome.isalpha():
    print('Insira apenas letras.')
    nome = input('Qual é o seu nome? ')

else:
    nome = str(nome)

#--------------------------------------------------------------------------------------------
SH = input('Quanto você recebe por hora? ') #Salário que ela recebe por hora

while not SH.isnumeric():
    print('Insira o valor em reais e centavos.')
    SH = input('Quanto você recebe por hora? ')
    
else:
    SH = int(SH)

#--------------------------------------------------------------------------------------------
HD = input('Quantas horas você pretende trabalhar por dia? ') #Horas de trabalho por dia

while not HD.isnumeric():
    print('Insira o valor em reais e centavos.')
    HD = input('Quanto você recebe por hora? ')
    
else:
    while not (int(HD) > 0 and int(HD) <= 24):
        print('Insira entre 0 e 24')
        HD = input()

HD = int(HD)
#--------------------------------------------------------------------------------------------
print('Pretendemos considerar 22 dias de trabalho no mês. Você está de acordo?')

DT = input('Insira 1 se concordar e 0 se não concordar: ') #Dias de trabalho por mês

S = 0

while not (DT == "0" or DT == "1"):
    print('Insira 1 se concordar e 0 se não concordar: ')
    DT = input()

if DT == "1":
    S = SH * HD * 22

elif DT == "0":
    print('Quantos dias serão trabalhados no mês? ')
    DT = input()
    while not (int(DT) > 0 and int(DT) <= 31):
        print('Insira entre 0 e 31')
        DT = input()
    
    S = SH * HD * int(DT)

print('Você seu merda filho da puta {} recebe {} por mês'.format(nome,S))
```

```python
#9. Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
#C = 5 * ((F-32) / 9).

F = input('Qual a temperatura em graus Fahrenheit? ')

while not F.isnumeric():
    F = input('Por favor, insira um valor numerico')

F = int(F)

print(type(F))

C = 5 * ((F - 32) / 9)

print('A temperatura em Celcius é:',C)
```

```python
#10. Faça um Programa que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.

C = input('Qual a temperatura em graus Celcius: ')

while not C.isnumeric():
    C = input('O valor de C deve ser um numero caralho: ')

C = int(C)

F = ((C / 5) * 9) + 32

print('A temperatura em Fahrenheit é:',F)
```

```python
#12. Tendo como dados de entrada a altura de uma pessoa, construa um algoritmo que calcule seu peso ideal,
# usando a seguinte fórmula: (72.7*altura) - 58

print('Vamos calcular o seu peso ideal')

h = input('Insira a sua altura: ')

while not h.isnumeric():
    h = input('Insira a sua altura: ')

h = float(h)

PI = (72.7 * h) - 58

print('O seu peso ideal é:',PI)
```
