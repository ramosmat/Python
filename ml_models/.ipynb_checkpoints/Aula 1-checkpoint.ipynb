```python
#Passo 4: Importar base de dados para o python
#pandas - pacote que permite importar base de dados em várias extensões como excel
import pandas as pd

tabela = pd.read_excel(r'C:\Users\Matheus\Desktop\DataScience\HashTag Python\Vendas - Dez.xlsx')
    #este comando busca arquivos de uma extensão especifica como: excel = arquivos excel, csv = arquivos csv
    #LEMBRAR DE SEMPRE INFORMAR A EXTENSÃO DO ARQUIVO
    
display(tabela)
```

```python
#Passo 5: Calcular os indicadores

#5.1 Faturamento = Soma da coluna Valor Final

faturamento = tabela['Valor Final'].sum() #.sum .count. average

#5.2 Quantidade de produtos = Soma da coluna Qunatidade

quantidade = tabela['Quantidade'].sum()

print('O faturamento é de =', faturamento)
print('A quantidade é de =', quantidade)
```
