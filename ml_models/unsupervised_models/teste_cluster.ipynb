```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

np.random.seed(42)  # Define a semente para reprodução dos resultados
data = np.concatenate([
    np.random.normal(5, 1, (100, 2)),   # Centro 1
    np.random.normal(10, 1, (100, 2)),  # Centro 2
    np.random.normal(15, 1, (100, 2)),  # Centro 3
])

kmeans = KMeans(n_clusters=3, n_init = 10)
kmeans.fit(data)

# Obter as coordenadas dos centros dos clusters
centroids = kmeans.cluster_centers_

# Obter os rótulos dos clusters para cada ponto de dado
labels = kmeans.labels_

# Plotar os pontos de dados e os centros dos clusters em um gráfico
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='D')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering')
plt.show()
```
