```python
#4. Nome na vertical em escada. Modifique o programa anterior de forma a mostrar o nome em formato de escada.

nome = input('Insira seu nome: ')

nome = nome.capitalize()

c = ''
print('')

for i in nome:
    c += i
    print(c)
    
print('\nO nome indicado foi:')
print(c)
```

```python
#5. Nome na vertical em escada invertida. Altere o programa anterior de modo que a escada seja invertida.

nome = input('Insira seu nome: ')

nome = nome.capitalize()

print('')

for i in range(0,len(nome)+1):
    print(nome[:len(nome)-i])
    
print('\nO nome indicado foi:')
print(nome)
```

```python
#7. Conta espaços e vogais. Dado uma string com uma frase informada pelo usuário (incluindo espaços em branco), conte:
    # quantos espaços em branco existem na frase.
    # quantas vezes aparecem as vogais a, e, i, o, u.
    
esp =' '

c_esp = 0
c_vog = 0

f = str(input('Insira uma frase: '))

f = f.lower()

for i in f:
    if i == esp:
        c_esp += 1
    
    elif i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u':
        c_vog += 1

print(f'\nA quantidade de espaços na frase é: {c_esp}.')
print(f'A quantidade de vogais na frase é: {c_vog}.')
```

```python
#8. Palíndromo. Um palíndromo é uma seqüência de caracteres cuja leitura é idêntica se feita da direita para esquerda ou
# vice−versa. Por exemplo: OSSO e OVO são palíndromos. Em textos mais complexos os espaços e pontuação são ignorados.
# A frase SUBI NO ONIBUS é o exemplo de uma frase palíndroma onde os espaços foram ignorados. Faça um programa que
# leia uma seqüência de caracteres, mostre−a e diga se é um palíndromo ou não.

f = str(input('Insira uma frase: '))

f = f.upper()

f = f.replace(' ','')

f_inv = f[::-1]

print(f'\n{f}')
print(f_inv)

if f == f_inv:
    print('\nMeu pau é pequeno seu filho da puta do caralho. É um políndromo.')

else:
    print('\nMeu pau continua pequeno seu merda. Não é um políndromo')
```

```python
#9. Verificação de CPF. Desenvolva um programa que solicite a digitação de um número de CPF no formato xxx.xxx.xxx-xx e
# indique se é um número válido ou inválido através da validação dos dígitos verificadores e dos caracteres de formatação.

cpf = input('Insira seu cpf: ')

while len(cpf) != 11:
    cpf = input('O cpf deve ter 11 caracteres: ')
    
print(f'O cpf é {cpf[:3]}.{cpf[3:6]}.{cpf[6:9]}-{cpf[9:11]}')
```

```python
#12. Valida e corrige número de telefone. Faça um programa que leia um número de telefone, e corrija o número no caso deste
# conter somente 7 dígitos, acrescentando o '3' na frente. O usuário pode informar o número com ou sem o traço separador.

num = input('Insira o numero de telefone: ')

if len(num) < 8:
    dif = 8 - len(num)
    nov = '3' * dif
    nov_num = nov + num
    print('O numero de telefone é:',nov_num[:4],'-',nov_num[4:])
    
else:
    print('O numero de telefone é:',num)
```

```python
#13. Jogo da palavra embaralhada. Desenvolva um jogo em que o usuário tenha que adivinhar uma palavra que será mostrada com as
# letras embaralhadas. O programa terá uma lista de palavras lidas de um arquivo texto e escolherá uma aleatoriamente.
# O jogador terá seis tentativas para adivinhar a palavra. Ao final a palavra deve ser mostrada na tela, informando se o usuário
# ganhou ou perdeu o jogo.

import random as rd

palavras = ['Menina','Terno','Licor','Lixeira','Forrado','Ovo','Perpendicular','Primordial','Pulsar','Sarda']

palavra = rd.choice(palavras)

#tratativas da palavra
palavra = palavra.strip() #apaga os espaçoes que tiver no começo ou final da string
palavra = palavra.upper()

n1 = palavra #Palavra não embaralhada

palavra = list(palavra)
rd.shuffle(palavra)

p = ''.join(palavra) #Palavra embaralhada // .join junta os itens da lista ou tupla com o que está dentro do apostrofe

c = 0

while True:
    print(f'Palavra embaralhada: {p}')
    n = input('Adivinhe a palavra que está embaralhada em 6 tentativas: ')
    n = n.upper()
    if n == n1:
        print('\nParabéns, você acertou!!')
        break
        
    elif c < 5:
        c += 1
        print(f'\nContinue, você já tentou {c} vezes.')
        
    else:
        print('\nVocê errou 6 vezes e perdeu o jogo.')
        break
```
